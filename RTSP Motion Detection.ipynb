```python
import numpy as np

def MotionTrack(f1, f2):
    
    #pythagorean distance between two frames
    f1_32 = np.float32(f1)
    f2_32 = np.float32(f2)
    diff32 = f1_32 - f2_32
    norm32 = np.sqrt(diff32[:,:,0]**2 + diff32[:,:,1]**2 + diff32[:,:,2]**2)/np.sqrt(255**2 + 255**2 + 255**2)
    motion = np.uint8(norm32*255)
    return motion
```

```python
import cv2
from datetime import datetime
import time

#Variable initialization
sdThresh = 10
font = cv2.FONT_HERSHEY_SIMPLEX
vcap = cv2.VideoCapture("rtsp://172.16.182.217:554/chID=2&streamType=Sub")
_, frame1 = vcap.read()
_, frame2 = vcap.read()
fps = vcap.get(cv2.CAP_PROP_FPS)   
timestamps = [vcap.get(cv2.CAP_PROP_POS_MSEC)]
fmt = "%Y-%m-%d %H:%M:%S %Z%z"


while(1):
    ret, frame = vcap.read()
    cv2.imshow('Raw', frame)
    cv2.waitKey(1)
    _, frame3 = vcap.read()
    rows, cols, _ = np.shape(frame3)
    dist = MotionTrack(frame1, frame3)
    
    frame1 = frame2
    frame2 = frame3
    
    mod = cv2.GaussianBlur(dist, (9,9), 0)
    _, thresh = cv2.threshold(mod, 100, 255, 0)
    _, stDev = cv2.meanStdDev(mod)
    
    #Shows how motion is detected visually. 
    cv2.imshow('Motion', mod)
    cv2.putText(frame2, "Standard Deviation - {}".format(round(stDev[0][0],0)), (70, 70), font, 1, (255, 0, 255), 1, cv2.LINE_AA)
    
    #Color/Punish
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV) 
    lower_dark = np.array([0,0,0]) 
    upper_dark = np.array([75,80,4])
    mask = cv2.inRange(hsv, lower_dark, upper_dark)
    
    
    #Time counter since initialization
    timestamps.append(vcap.get(cv2.CAP_PROP_POS_MSEC))

    for i, ts in enumerate(timestamps):
        frameDif = abs(ts) # ts - cts, is cts necessary? 
        mins = frameDif/60000
        minutes = int(mins)
        seconds = int((mins - minutes)*60)
        if seconds == 0 or 1 or 2 or 3 or 4 or 5 or 6 or 7 or 8 or 9:
            seconds = str(seconds).zfill(2)
        
    
    # Signals when motion occurs 
    if stDev > sdThresh:
        now = datetime.now()
        begin = time.time()
        print("Bird moved at: " + str(now.strftime(fmt)))
        print("Bird moved " + str(minutes) + ':' + str(seconds) + '(m:s) into initialization')
        print()
        
    #Signals when lights are shut off
    if cv2.countNonZero(mask) == 0:
        nowp = (datetime.now())
        beginp = time.time()
        print('Bird is punished at: ' + str(nowp.strftime(fmt)))
        print("Bird punished " + str(minutes) + ':' + str(seconds) + '(m:s) into initialization')
        print()
            
    

vcap.release()
```
